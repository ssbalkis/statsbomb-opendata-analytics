# WORLD CUP 2022 FINAL MATCH ANALYTICS

```python
#importing necessary libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from mplsoccer import VerticalPitch, Pitch, Sbopen
```

```python
#parsing
sb = Sbopen()
```

```python
#looking for all competitons
competitions = sb.competition()
competitions.info()
print("\n")
competitions
```

```python
#looking for matches in selected competition and season
matches = sb.match(competition_id = 43, season_id = 106)
matches.info()
print("\n")
matches
```

```python
#selected match_id
match = 3869685
```

```python
#looking lineups
lineups = sb.lineup(match)
lineups.info()
print("\n")
lineups
```

```python
#what happened in match...
events, related, freeze, tactics = sb.event(match)

events.info()
print("\n")
related.info()
print("\n")
freeze.info()
print("\n")
tactics.info()
```

```python
#get team names
team1, team2 = events.team_name.unique()
team1, team2
```

## Plotting Shots

```python
#dataframe of shots
shots = events.loc[events['type_name'] == 'Shot'].set_index('id')
shots.info()
print("\n")
shots
```

```python
#create pitch
pitch = Pitch(line_color='black')
fig, ax = pitch.grid(grid_height=0.9, title_height=0.06, axis=False,
                     endnote_height=0.04, title_space=0, endnote_space=0)

#query
mask_argentina = (events.type_name == 'Shot') & (events.team_name == team1)
mask_france = (events.type_name == 'Shot') & (events.team_name == team2)

#finding rows in the events and keeping only necessary columns
events_argentina = events.loc[mask_argentina, ['x', 'y', 'outcome_name', "player_name"]]
events_france = events.loc[mask_france, ['x', 'y', 'outcome_name', "player_name"]]


#plot them - if shot ended with Goal - alpha 1 and add name
#for argentina
for i, row in events_argentina.iterrows():
    if row["outcome_name"] == 'Goal':
    #make circle
       pitch.scatter(row.x, row.y, alpha = 1, s = 500, color = "grey", ax=ax['pitch'])
       pitch.annotate(row["player_name"], (row.x + 1, row.y - 2), ax=ax['pitch'], fontsize = 12)
    else:
       pitch.scatter(row.x, row.y, alpha = 0.2, s = 500, color = "grey", ax=ax['pitch'])

#for france we need to revert coordinates
for i, row in events_france.iterrows():
    if row["outcome_name"] == 'Goal':
       pitch.scatter(120 - row.x, 80 - row.y, alpha = 1, s = 500, color = "blue", ax=ax['pitch'])
       pitch.annotate(row["player_name"], (120 - row.x + 1, 80 - row.y - 2), ax=ax['pitch'], fontsize = 12)
    else:
       pitch.scatter(120 - row.x, 80 - row.y, alpha = 0.2, s = 500, color = "blue", ax=ax['pitch'])

#figure header
fig.suptitle("Argentina (Grey) and France (Blue) shots", fontsize = 30)
plt.show()
```

```python
#only france shots on half pitch
pitch = VerticalPitch(line_color='black', half = True)
fig, ax = pitch.grid(grid_height=0.9, title_height=0.06, axis=False,
                     endnote_height=0.04, title_space=0, endnote_space=0)
#plotting all shots
pitch.scatter(events_france.x, events_france.y, alpha = 1, s = 500, color = "blue", ax=ax['pitch'], edgecolors="black")
fig.suptitle("France shots against Argentina", fontsize = 30)
plt.show()
```

```python
#prepare the dataframe of passes by Argentina that were no-throw ins
mask_argentina = (events.type_name == 'Pass') & (events.team_name == "Argentina") & (events.sub_type_name != "Throw-in")
argentina_passes = events.loc[mask_argentina, ['x', 'y', 'end_x', 'end_y', 'player_name']]

#get the list of all players who made a pass
names = argentina_passes['player_name'].unique()

#draw 4x4 pitches
pitch = Pitch(line_color='black', pad_top=20)
fig, axs = pitch.grid(ncols = 4, nrows = 4, grid_height=0.85, title_height=0.06, axis=False,
                     endnote_height=0.04, title_space=0.04, endnote_space=0.01)

#for each player
for name, ax in zip(names, axs['pitch'].flat[:len(names)]):
    #put player name over the plot
    ax.text(60, -10, name,
            ha='center', va='center', fontsize=8)
    #take only passes by this player
    who_pass = argentina_passes.loc[argentina_passes["player_name"] == name]
    #scatter
    pitch.scatter(who_pass.x, who_pass.y, alpha = 0.2, s = 50, color = "blue", ax=ax)
    #plot arrow
    pitch.arrows(who_pass.x, who_pass.y,
            who_pass.end_x, who_pass.end_y, color = "blue", ax=ax, width=1)

#We have more than enough pitches - remove them
for ax in axs['pitch'][-1, 16 - len(names):]:
    ax.remove()

#Another way to set title using mplsoccer
axs['title'].text(0.5, 0.5, 'Argentina passes against France', ha='center', va='center', fontsize=30)
plt.show()
```

## Passing Networks

```python
#check for index of first sub
sub = events.loc[events["type_name"] == "Substitution"].loc[events["team_name"] == "Argentina"].iloc[0]["index"]

#make df with successfull passes by England until the first substitution
mask_argentina = (events.type_name == 'Pass') & (events.team_name == "Argentina") & (events.index < sub) & (events.outcome_name.isnull()) & (events.sub_type_name != "Throw-in")

#taking necessary columns
argentina_passes = events.loc[mask_argentina, ['x', 'y', 'end_x', 'end_y', "player_name", "pass_recipient_name"]]

#adjusting that only the surname of a player is presented.
argentina_passes["player_name"] = argentina_passes["player_name"].apply(lambda x: str(x).split()[-1])
argentina_passes["pass_recipient_name"] = argentina_passes["pass_recipient_name"].apply(lambda x: str(x).split()[-1])
```

```python
scatter_df = pd.DataFrame()
for i, name in enumerate(argentina_passes["player_name"].unique()):
    passx = argentina_passes.loc[argentina_passes["player_name"] == name]["x"].to_numpy()
    recx = argentina_passes.loc[argentina_passes["pass_recipient_name"] == name]["end_x"].to_numpy()
    passy = argentina_passes.loc[argentina_passes["player_name"] == name]["y"].to_numpy()
    recy = argentina_passes.loc[argentina_passes["pass_recipient_name"] == name]["end_y"].to_numpy()
    scatter_df.at[i, "player_name"] = name
    #make sure that x and y location for each circle representing the player is the average of passes and receptions
    scatter_df.at[i, "x"] = np.mean(np.concatenate([passx, recx]))
    scatter_df.at[i, "y"] = np.mean(np.concatenate([passy, recy]))
    #calculate number of passes
    scatter_df.at[i, "no"] = argentina_passes.loc[argentina_passes["player_name"] == name].count().iloc[0]

#adjust the size of a circle so that the player who made more passes
scatter_df['marker_size'] = (scatter_df['no'] / scatter_df['no'].max() * 1500)
```

```python
#counting passes between players
argentina_passes["pair_key"] = argentina_passes.apply(lambda x: "_".join(sorted([x["player_name"], x["pass_recipient_name"]])), axis=1)
lines_df = argentina_passes.groupby(["pair_key"]).x.count().reset_index()
lines_df.rename({'x':'pass_count'}, axis='columns', inplace=True)
#setting a treshold. You can try to investigate how it changes when you change it.
lines_df = lines_df[lines_df['pass_count']>2]
```

```python
#Drawing pitch
pitch = Pitch(line_color='black')
fig, ax = pitch.grid(grid_height=0.9, title_height=0.06, axis=False,
                     endnote_height=0.04, title_space=0, endnote_space=0)
#Scatter the location on the pitch
pitch.scatter(scatter_df.x, scatter_df.y, s=scatter_df.marker_size, color='skyblue', edgecolors='steelblue', linewidth=1, alpha=1, ax=ax["pitch"], zorder = 3)
#annotating player name
for i, row in scatter_df.iterrows():
    pitch.annotate(row.player_name, xy=(row.x, row.y), c='black', va='center', ha='center', weight = "bold", size=16, ax=ax["pitch"], zorder = 4)

fig.suptitle("Nodes location - Argentina", fontsize = 30)
plt.show()
```

```python
#plot once again pitch and vertices
pitch = Pitch(line_color='black')
fig, ax = pitch.grid(grid_height=0.9, title_height=0.06, axis=False,
                     endnote_height=0.04, title_space=0, endnote_space=0)
pitch.scatter(scatter_df.x, scatter_df.y, s=scatter_df.marker_size, color='skyblue', edgecolors='steelblue', linewidth=1, alpha=1, ax=ax["pitch"], zorder = 3)
for i, row in scatter_df.iterrows():
    pitch.annotate(row.player_name, xy=(row.x, row.y), c='black', va='center', ha='center', weight = "bold", size=16, ax=ax["pitch"], zorder = 4)

for i, row in lines_df.iterrows():
        player1 = row["pair_key"].split("_")[0]
        player2 = row['pair_key'].split("_")[1]
        #take the average location of players to plot a line between them
        player1_x = scatter_df.loc[scatter_df["player_name"] == player1]['x'].iloc[0]
        player1_y = scatter_df.loc[scatter_df["player_name"] == player1]['y'].iloc[0]
        player2_x = scatter_df.loc[scatter_df["player_name"] == player2]['x'].iloc[0]
        player2_y = scatter_df.loc[scatter_df["player_name"] == player2]['y'].iloc[0]
        num_passes = row["pass_count"]
        #adjust the line width so that the more passes, the wider the line
        line_width = (num_passes / lines_df['pass_count'].max() * 10)
        #plot lines on the pitch
        pitch.lines(player1_x, player1_y, player2_x, player2_y,
                        alpha=1, lw=line_width, zorder=2, color="salmon", ax = ax["pitch"])

fig.suptitle("Argentina Passing Network against France", fontsize = 30)
plt.show()
```

```python
#calculate number of successful passes by player
no_passes = argentina_passes.groupby(['player_name']).x.count().reset_index()
no_passes.rename({'x':'pass_count'}, axis='columns', inplace=True)
#find one who made most passes
max_no = no_passes["pass_count"].max()
#calculate the denominator - 10*the total sum of passes
denominator = 10*no_passes["pass_count"].sum()
#calculate the nominator
nominator = (max_no - no_passes["pass_count"]).sum()
#calculate the centralisation index
centralisation_index = nominator/denominator
print("Centralisation index is ", centralisation_index)
```
